```python
#load required libraries
import numpy as np
import cv2
import matplotlib.pyplot as plt

from tensorflow.keras.applications import resnet50
from tensorflow.keras.preprocessing import image

import sys
sys.path.append('..')

from src.gradcam import get_grad_cam, overlay_heatmap
```

```python
#create resent50 model and load imagenet weights
model = resnet50.ResNet50(weights='imagenet', include_top=True)
```

```python
#Load sample image
img = image.load_img('../sample_images/cat.jpg', target_size=(224, 224))
img = image.img_to_array(img)
#display the image
plt.subplot(1, 2, 1)
plt.imshow(img/255.0)
plt.title('Image')
plt.xticks([])
plt.yticks([])

#predict the class
x = np.expand_dims(img, axis=0)
x = resnet50.preprocess_input(x)
preds = model.predict(x)
print("Top3 predictions: {}".format(resnet50.decode_predictions(preds, top=3)[0]))


#get the grad cam 
i = np.argmax(preds)
heatmap  = get_grad_cam(x, model, 'conv5_block3_out', i)
#overlay heatmap on the input image
output = overlay_heatmap(img, heatmap)


#display the output
plt.subplot(1, 2, 2)
plt.imshow(output)
plt.title('grad CAM')
plt.xticks([])
plt.yticks([])
plt.show()
```
